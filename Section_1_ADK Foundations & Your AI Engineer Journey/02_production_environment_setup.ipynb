## Step 5: Create Virtual Environment ğŸ 

**In VS Code Terminal (recommended) or Command Prompt:**

### Open Terminal in VS Code:
- **Menu:** Terminal â†’ New Terminal
- **Shortcut:** `Ctrl+`` (backtick) or `Cmd+`` (Mac)

### Create and Activate Virtual Environment:
```bash

# macOS/Linux:

```

**âœ… Success indicator:** Your terminal prompt should show `(venv)` at the beginning

### Verify Virtual Environment:

````python
# Check virtual environment status
import sys
from pathlib import Path

print(f"ğŸ Python Executable: {sys.executable}")
print(f"ğŸ“¦ Python Prefix: {sys.prefix}")

# Check if we're in a virtual environment
in_venv = hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)

if in_venv:
    print("\nâœ… Virtual environment is ACTIVE! ğŸ‰")
    
    # Check if it's our project venv
    if 'venv' in sys.executable:
        print("âœ… Using project virtual environment")
    else:
        print("ğŸ’¡ Using a virtual environment (may not be project-specific)")
        
else:
    print("\nâš ï¸ Virtual environment not detected")
    print("ğŸ’¡ Run these commands in VS Code terminal:")
    print("   1. python -m venv venv")
    print("   2. Windows: venv\\Scripts\\activate")
    print("   2. Mac/Linux: source venv/bin/activate")

# Check for venv folder
venv_path = Path('venv')
if venv_path.exists():
    print(f"\nğŸ“ Virtual environment folder exists: âœ…")
else:
    print("\nğŸ“ Virtual environment folder not found")
    print("ğŸ’¡ Create it with: python -m venv venv")
````

### Verify API Key Setup:

````python
# Check environment configuration
import os
from pathlib import Path

print("ğŸ” Environment Configuration Check")
print("=" * 40)

# Check if .env file exists
env_file = Path('.env')
if env_file.exists():
    print("âœ… .env file found")
    
    # Try to load environment variables
    try:
        from dotenv import load_dotenv
        load_dotenv()
        
        # Check Google API key (without exposing it)
        api_key = os.getenv('GOOGLE_API_KEY')
        if api_key and len(api_key) > 20:
            print(f"âœ… Google API key configured")
            print(f"   Key preview: {api_key[:8]}...{api_key[-4:]}")
            
            # Check default model
            model = os.getenv('DEFAULT_MODEL', 'gemini-2.0-flash')
            print(f"âœ… Default model: {model}")
            
            print("\nğŸ‰ API key configuration complete!")
            
        else:
            print("âŒ Google API key not configured or too short")
            print("ğŸ’¡ Edit .env file and add your API key")
            print("ğŸ’¡ Get key from: https://aistudio.google.com/app/apikey")
            
    except ImportError:
        print("âš ï¸ python-dotenv not installed yet (will install in next step)")
        print("ğŸ’¡ We'll verify this again after installing dependencies")
        
else:
    print("âŒ .env file not found")
    print("ğŸ’¡ Copy .env.example to .env first")
    print("ğŸ’¡ Then edit .env and add your Google API key")

print("\nğŸ”’ Security note: .env file is protected by .gitignore")
````

### Verify Installation:

````python
# Verify Google ADK installation
print("ğŸ“¦ Google ADK Installation Verification")
print("=" * 45)

# Core dependencies to check
dependencies = {
    'google.adk': 'Google Agent Development Kit',
    'google.generativeai': 'Google AI integration',
    'dotenv': 'Environment variable management',
    'pydantic': 'Data validation',
    'asyncio': 'Asynchronous programming'
}

installed_count = 0
total_deps = len(dependencies)

for package, description in dependencies.items():
    try:
        if package == 'google.adk':
            from google.adk.agents import Agent
        elif package == 'dotenv':
            from dotenv import load_dotenv
        elif package == 'asyncio':
            import asyncio
        else:
            __import__(package.replace('-', '_'))
        
        print(f"âœ… {package:<20} - {description}")
        installed_count += 1
    except ImportError:
        print(f"âŒ {package:<20} - {description} (NOT INSTALLED)")

print(f"\nğŸ“Š Installation Status: {installed_count}/{total_deps} packages")

if installed_count == total_deps:
    print("ğŸ‰ ALL DEPENDENCIES INSTALLED!")
    print("ğŸš€ Google ADK development environment ready")
elif installed_count >= 3:
    print("âœ… Core packages installed - mostly ready")
else:
    print("âš ï¸ Installation incomplete")
    print("ğŸ’¡ Run in terminal: pip install -r requirements.txt")
    print("ğŸ’¡ Make sure virtual environment is activated (venv)")
````
