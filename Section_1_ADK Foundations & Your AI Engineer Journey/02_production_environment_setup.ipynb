````python
# Phase 1 Completion Verification
print("üîç PHASE 1 COMPLETION CHECK")
print("=" * 40)

import sys
from pathlib import Path

# Check Python version
version = sys.version_info
print(f"üêç Python Version: {version.major}.{version.minor}.{version.micro}")

if version.major == 3 and version.minor >= 9 and version.minor <= 11:
    print("‚úÖ Python version is perfect for Google ADK!")
elif version.minor == 12:
    print("‚ö†Ô∏è Python 3.12 may have compatibility issues")
    print("üí° Consider using Python 3.11 for best results")
elif version.minor < 9:
    print("‚ùå Python version too old")
    print("üí° Install Python 3.9+ from python.org")
    print("üõë STOP HERE - Update Python first")
else:
    print(f"‚úÖ Python {version.major}.{version.minor} should work")

# Check virtual environment
in_venv = hasattr(sys, 'real_prefix') or (hasattr(sys, 'base_prefix') and sys.base_prefix != sys.prefix)
print(f"\nüè† Virtual Environment: {'‚úÖ Active' if in_venv else '‚ùå Not Active'}")

if in_venv:
    if 'venv' in sys.executable:
        print("‚úÖ Using project virtual environment")
    else:
        print("‚ö†Ô∏è Using virtual environment (may not be project-specific)")
    print(f"üìç Python executable: {sys.executable}")
else:
    print("‚ùå No virtual environment detected")
    print("üí° Complete Phase 1 Step 4: Create and activate virtual environment")
    print("üí° Then restart this notebook kernel")

# Check if we're in the right directory
current_dir = Path.cwd()
print(f"\nüìÇ Current Directory: {current_dir.name}")

# Look for course files
expected_files = ['requirements.txt', '.env.example', 'README.md']
found_files = []
missing_files = []

print("\nüìÑ Project Files:")
for file in expected_files:
    if Path(file).exists():
        found_files.append(file)
        print(f"‚úÖ {file} found")
    else:
        missing_files.append(file)
        print(f"‚ùå {file} missing")

# Final Phase 1 assessment
print(f"\nüìä Phase 1 Status: {len(found_files)}/{len(expected_files)} requirements met")

if len(found_files) == len(expected_files) and version.minor >= 9 and in_venv:
    print("üéâ PHASE 1 COMPLETE! Ready for verification")
    print("‚ñ∂Ô∏è Continue to next cell")
else:
    print("üõë PHASE 1 INCOMPLETE")
    if missing_files:
        print(f"üí° Missing files: {missing_files}")
        print("üí° Make sure you're in the google-adk-enterprise-agents folder")
    if not in_venv:
        print("üí° Virtual environment not active")
    print("\nüîÑ Complete Phase 1 setup before continuing")
````

````python
# Verify Dependencies Installation
print("üì¶ DEPENDENCY VERIFICATION")
print("=" * 40)

# Core dependencies to check
dependencies = {
    'google.adk': 'Google Agent Development Kit',
    'google.generativeai': 'Google AI integration',
    'dotenv': 'Environment variable management',
    'pydantic': 'Data validation',
    'asyncio': 'Asynchronous programming'
}

installed_count = 0
total_deps = len(dependencies)

print("üîç Testing imports...\n")

for package, description in dependencies.items():
    try:
        if package == 'google.adk':
            from google.adk.agents import Agent
        elif package == 'dotenv':
            from dotenv import load_dotenv
        elif package == 'asyncio':
            import asyncio
        else:
            __import__(package.replace('-', '_'))
        
        print(f"‚úÖ {package:<20} - {description}")
        installed_count += 1
    except ImportError as e:
        print(f"‚ùå {package:<20} - {description} (NOT INSTALLED)")
        print(f"   Error: {str(e)[:60]}...")

print(f"\nüìä Installation Status: {installed_count}/{total_deps} packages")

if installed_count == total_deps:
    print("üéâ ALL DEPENDENCIES INSTALLED!")
    print("üöÄ Google ADK development environment ready")
    print("‚ñ∂Ô∏è Continue to environment configuration")
elif installed_count >= 3:
    print("‚úÖ Core packages installed - mostly ready")
    print("üí° You can proceed, some optional packages may need attention")
else:
    print("‚ùå Critical packages missing")
    print("\nüîß TROUBLESHOOTING:")
    print("1. Ensure virtual environment is activated in terminal")
    print("2. Run in VS Code terminal: pip install -r requirements.txt")
    print("3. Restart Jupyter kernel after installation")
    print("4. Make sure you selected 'Google ADK Enterprise' kernel")
    print("\nüí° If still failing, complete Phase 1 setup steps first")
````

````python
# Environment Configuration Check
print("üîê ENVIRONMENT CONFIGURATION")
print("=" * 40)

import os
from pathlib import Path

env_file = Path('.env')
env_example = Path('.env.example')

print("üìÑ Environment Files:")
if env_example.exists():
    print("‚úÖ .env.example template found")
else:
    print("‚ùå .env.example template missing")

if env_file.exists():
    print("‚úÖ .env configuration file found")
    
    # Try to load and check the API key
    try:
        from dotenv import load_dotenv
        load_dotenv()
        
        # Check Google API key (without exposing it)
        api_key = os.getenv('GOOGLE_API_KEY')
        if api_key and api_key != 'your_google_api_key_here' and len(api_key) > 30:
            print(f"‚úÖ Google API key configured")
            print(f"   Key preview: {api_key[:8]}...{api_key[-4:]}")
            
            # Check default model
            model = os.getenv('DEFAULT_MODEL', 'gemini-2.0-flash')
            print(f"‚úÖ Default model: {model}")
            
            print("\nüéâ API key configuration complete!")
            
        else:
            print("‚ö†Ô∏è Google API key not configured properly")
            print("\nüîß TO FIX:")
            print("1. Open .env file in VS Code")
            print("2. Get API key from: https://aistudio.google.com/app/apikey")
            print("3. Replace 'your_google_api_key_here' with your actual key")
            print("4. Save file and re-run this cell")
            
    except ImportError:
        print("‚ùå python-dotenv not installed")
        print("üí° Install dependencies first: pip install -r requirements.txt")
        
else:
    print("‚ùå .env configuration file missing")
    print("\nüîß TO CREATE:")
    if env_example.exists():
        print("1. In VS Code terminal, run:")
        print("   Windows: copy .env.example .env")
        print("   Mac/Linux: cp .env.example .env")
        print("2. Edit .env file and add your Google API key")
        print("3. Re-run this cell")
    else:
        print("1. Create .env file manually")
        print("2. Add: GOOGLE_API_KEY=your_actual_key_here")
        print("3. Get key from: https://aistudio.google.com/app/apikey")

print("\nüîí Security Note: .env file is protected by .gitignore")
print("üîí Your API key will never be committed to version control")
````
